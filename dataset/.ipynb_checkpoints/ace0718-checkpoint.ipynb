```python
import math
import numpy as np  # 引入numpy、pandas库
import pandas as pd
from dbfread import DBF
from pandas import DataFrame
from IPython.display import display
pd.options.display.max_columns = None
import matplotlib.pyplot as plt

import seaborn as sns


```

```python


path_1 = "./IBTrACS.SP.list.v04r00.lines.dbf"



df_raw1 = pd.DataFrame(iter(DBF(path_1)))
```

```python
def max_wind_sum_of_square(ls):
    nums = 0
    for tmp in ls:
        if np.isnan(tmp) == True:
            continue
        else:
            nums += tmp**2
    return nums
```

```python
df = df_raw1[(df_raw1['USA_STATUS'] != "EX") 
            & (df_raw1['USA_STATUS'] != "SS") 
            & (df_raw1['USA_STATUS'] != "PT") 
            & (df_raw1['USA_STATUS'] != "IN")
            & (df_raw1['USA_STATUS'] != "DS") 
            & (df_raw1['USA_STATUS'] != "WV")
            & (df_raw1['USA_STATUS'] != "ET") 
            & (df_raw1['USA_STATUS'] != "MD") 
            & (df_raw1['USA_STATUS'] != "XX")]# 去掉表格属性USA_STATUS为EX的数据
```

```python
df_raw1.shape
```

```python
df.shape
```

```python
df = df.loc[df['USA_WIND'] > 35]
```

```python
df.shape
```

```python
df_raw1.head(100)
```

```python
col_year = []
col_accumulated_indensity = []
for year in range(1982, 2022):
    df_typh_year = df[df["SEASON"] == year]

    df_typh_year = df_typh_year.loc[df_typh_year['USA_WIND'] > 35] # 筛选风速大于等于35台风

#     df_typh = df_nonNan_moreThan_34[df_nonNan_moreThan_34["SEASON"] == year]
    
    typhone_times = df_typh_year['NUMBER'].value_counts().sort_index().reset_index()  # 统计每个台风出现的次数 
    sum_each_year_indensity = 0 # 一年的ace和

    for typhone_number in typhone_times['index']:  # 遍历台风的编号
        df_typh_number = df_typh_year[df_typh_year["NUMBER"] == typhone_number]  # 圈出台风编号为typhone_number的表格
#         print("NO.%d typhone max: %.1f!"%(typhone_number, df_typh_number["USA_WIND"].max()))
        typh_times = int(typhone_times[typhone_times['index'] == typhone_number]['NUMBER'])      

        ls_wind_indensity = list(df_typh_number["USA_WIND"])
        indesity_length = len(ls_wind_indensity) # 风速强度列表长度
        if typh_times % 2 == 0:  # 出现偶数次台风
#             print("The No. %d typhone is even and detection times are %d."%(typhone_number, typh_times))
    #         ls_wind_indensity = list(df_typh_number["USA_WIND"])
    #         indesity_length = len(ls_wind_indensity) # 风速强度列表长度
            typh_double_line_max = [] # 两行最大值列表
            for i in range(0, indesity_length, 2):  # 每两行遍历
                tmp = max(ls_wind_indensity[i], ls_wind_indensity[i+1]) # 取两行中最大的值
                typh_double_line_max.append(tmp) 
            sum_after_square = max_wind_sum_of_square(typh_double_line_max)
#             print("The sum after square is %.2f"%sum_after_square)

        elif typh_times % 2 == 1:  # 出现奇数次台风
#             print("The No. %d typhone is odd and detection times are %d."%(typhone_number, typh_times))
    #         ls_wind_indensity = list(df_typh_number["USA_WIND"])
    #         indesity_length = len(ls_wind_indensity) # 风速强度列表长度
            typh_double_line_max = [] # 两行最大值列表
            for i in range(0, indesity_length - 1, 2):  # 每两行遍历
                tmp = max(ls_wind_indensity[i], ls_wind_indensity[i+1]) # 取两行中最大的值
                typh_double_line_max.append(tmp)

            last_line = (ls_wind_indensity[indesity_length-1] ** 2) * 0.5
            sum_after_square = max_wind_sum_of_square(typh_double_line_max) + last_line
#             print("The sum after square is %.2f"%sum_after_square)
        sum_each_year_indensity += sum_after_square
    
    col_year.append(year)
    col_accumulated_indensity.append(sum_each_year_indensity)
    print("The %d year ace is %.2e"%(year, sum_each_year_indensity))

ace = pd.DataFrame({'Year': col_year, 'ACE': col_accumulated_indensity})
```

```python
ace.to_excel("./ACE SP.xlsx")  # 存储到excel中
```

```python
ace.plot(x='Year')
ace.set_index('Year')
plt.show()
```
