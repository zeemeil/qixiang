```python
import numpy as np  # 引入numpy、pandas库
import pandas as pd
from dbfread import DBF
from pandas import DataFrame
from IPython.display import display
pd.options.display.max_columns = None

path = "./SI.dbf"

def classification(typh_intensity):
    if typh_intensity >= 1 and typh_intensity <= 34:
        result = 'D'
    elif typh_intensity > 34 and typh_intensity <= 63:
        result = 'S'
    elif typh_intensity > 63 and typh_intensity <= 82:
        result = 'C1'
    elif typh_intensity > 82 and typh_intensity <= 95:
        result = 'C2'
    elif typh_intensity > 95 and typh_intensity <= 112:
        result = 'C3'
    elif typh_intensity > 112 and typh_intensity <= 136:
        result = 'C4'        
    elif typh_intensity > 136:
        result = 'C5'
    else:
        result = "WRONG"
    return result

df_raw = pd.DataFrame(iter(DBF(path)))

df = df_raw
```

```python
df_drop_usa_state = df[(df['USA_STATUS'] != "EX") 
            & (df['USA_STATUS'] != "SS") 
            & (df['USA_STATUS'] != "PT") 
            & (df['USA_STATUS'] != "IN")
            & (df['USA_STATUS'] != "DS") 
            & (df['USA_STATUS'] != "WV")
            & (df['USA_STATUS'] != "ET") 
            & (df['USA_STATUS'] != "MD") 
            & (df['USA_STATUS'] != "XX")]# 去掉表格属性USA_STATUS为EX....的数据
```

```python
df_MoreThan35 = df_drop_usa_state[(df_drop_usa_state['USA_WIND'] >= 35)]
```

```python
for year in range(1976, 2022):
    df_year2next = df_MoreThan35[((df_MoreThan35['year'] == year) & (df_MoreThan35['month'] >= 7)) 
   | ((df_MoreThan35['year'] == year+1) & (df_MoreThan35['month'] <= 6))]
    day = df_year2next.shape[0] * 3 / 24
    print("The %d year 7 month to %d year 6 month typhone has %.3f days!"%(year, year+1, day))
```

```python
# df_typh = df[df["SEASON"] == 1982]

each_year_days = []
each_year_S_days = []
each_year_C_days = []

for year in range(1982, 2022):
    df_typh = df[df["SEASON"] == year] 
    typhone_times = df_typh['NUMBER'].value_counts()  # 统计每个台风出现的次数 
    number_typhone = typhone_times[:].index[:].tolist() # 统计台风的编号
    number_typhone.sort() # 对编号进行排序
    ls_max_intensity = [] # 创建最大台风强度 空列表
    ls_classification = []  # 创建台风分类列表
    for index in range(len(number_typhone)):
        df5 = df_typh[df_typh["NUMBER"] == number_typhone[index]]
        index_max = df5['USA_WIND'].max()
        classifation_result = classification(index_max)
        ls_classification.append(classifation_result) 
        ls_max_intensity.append(index_max)

    level_number = list(zip(ls_classification, number_typhone)) # 台风编号的级别
    need_level_number = [] # 需要的等级台风的编号
    for i in range(len(level_number)):
        tmp = level_number[i]
        if tmp[0] in ['S', 'C1', 'C2', 'C3', 'C4', 'C5']:
            need_level_number.append(tmp)

    accumulate_days = 0
    S_level_days = 0 # S级别风
    C_level_days = 0 # C1-C5级别风

    for item in need_level_number:
        typh_number = df_typh.loc[df_typh['NUMBER'] == item[1]]  # 台风item[1]出现的列数
        typh_times = typh_number['NUMBER'].value_counts()  # 台风item[1]出现的次数
    #     print(item[0], typh_times)
        if item[0] in ['S']:
            S_typh_days = (typh_times.values[0] - 1)*3/24  # 台风item[1]现了多少小时
            S_level_days += S_typh_days
        else:
            C_typh_days = (typh_times.values[0] - 1)*3/24  # 台风item[1]现了多少小时
            C_level_days += C_typh_days

#     print("The S level typhone days are %.3f.\nThe all C level typhone are %.3f."%(S_level_days, C_level_days))
#     print("====The whole days are %.3f!====="%(S_level_days+C_level_days))
    each_year_S_days.append(S_level_days)
    each_year_C_days.append(C_level_days)
    each_year_days.append(S_level_days+C_level_days)

    print("The %d's typhone has %.4f days!"%(year, S_level_days+C_level_days))
       
print(each_year_S_days)
print(each_year_C_days)
print(each_year_days)
```

```python
data = {'year':[x for x in range(1982, 2022)],
       'S': each_year_S_days,
       'C': each_year_C_days,
       'All':each_year_days
       }
result = pd.DataFrame(data)

result.to_excel("./DSP.xlsx")
```
