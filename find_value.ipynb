```python
from scipy.optimize import fsolve

def func(v):
    x, = v.tolist()
    return [
        361 * (x ** 4) + 58 * (x ** 2) - 1530 * x -115700
    ]

r = fsolve(func, [5])
print(r[0])
r = fsolve(func, [-10])
print(r[0])
```

```python
import sympy as sp
import math
import numpy as np 
x = sp.Symbol('x')
#x = sp.Symbol('x',real='True') # If only real number roots are required
f_X = 361 * (x ** 4) + 58 * (x ** 2) - 1530 * x -115700
root_List = sp.solve(f_X)

print("The 1 result is ",root_List[0].evalf(),",")
print("The 2 result is ",root_List[1].evalf(),",")
print("The 3 result is ",root_List[2].evalf(),",")
print("The 4 result is ",root_List[3].evalf(),".")
```
