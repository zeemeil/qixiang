```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 读入数据
data = pd.read_csv(r'./data.csv', encoding='GBK')
X = data.drop('target',axis=1)
y = data['target']
```

```python
X.shape
```

```python
X.describe()
```

```python
X.isnull().sum()
```

```python
# 计算每个变量的缺失率
missing_percentages = X.isnull().mean()

# 筛选出缺失率在 80% 以上的变量
high_missing_variables = missing_percentages[missing_percentages > 0.8].index

# 删除这些变量
X = X.drop(high_missing_variables, axis=1)
```

```python
 X.mean()
```

```python
# 计算每一列的均值
means = X.mean()

# 填补缺失值
X.fillna(means, inplace=True)
```

```python
# 遍历每个分类变量进行one-hot编码
for column in X.select_dtypes(include=['object']):
    encoded_columns = pd.get_dummies(X[column], prefix=column)
    X = pd.concat([X, encoded_columns], axis=1)
    X.drop(column, axis=1, inplace=True)
# 打印编码后的数据
print(X)
```

```python
# 将'Idx'列设置为索引列
X = X.set_index('Idx')

# 打印设置后的DataFrame
print(X)
```

```python
from sklearn.preprocessing import StandardScaler
# 创建StandardScaler对象
scaler = StandardScaler()

# 对自变量进行标准化
X_scaled = scaler.fit_transform(X)

# 创建标准化后的DataFrame
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

# 打印标准化后的自变量
print(X_scaled_df)
```

```python
from sklearn.decomposition import PCA
# 创建PCA对象
pca = PCA(n_components=0.80)  # 设置要保留的主成分个数


# 对自变量进行主成分分析
X_pca = pca.fit_transform(X_scaled_df)

# 获取保留的主成分个数
n_components = pca.n_components_

# 创建主成分分析后的DataFrame
columns = ['PC{}'.format(i+1) for i in range(n_components)]
X_pca_df = pd.DataFrame(X_pca, columns=columns)

# 打印主成分分析后的结果
print(X_pca_df)
```

```python
from sklearn.preprocessing import MinMaxScaler

# 创建一个MinMaxScaler对象
scaler = MinMaxScaler()

# 对数据进行缩放
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))

fpr, tpr, thresholds = roc_curve(y_test, knn.predict_proba(X_test)[:,1])
roc_auc = auc(fpr, tpr)
print('AUC:', roc_auc)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=1)
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))

fpr, tpr, thresholds = roc_curve(y_test, dt.predict_proba(X_test)[:,1])
roc_auc = auc(fpr, tpr)
print('AUC:', roc_auc)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))

fpr, tpr, thresholds = roc_curve(y_test, lr.predict_proba(X_test)[:,1])
roc_auc = auc(fpr, tpr)
print('AUC:', roc_auc)
```

```python
from sklearn.metrics import roc_curve, auc

# KNN的AUC值
fpr, tpr, thresholds = roc_curve(y_test, knn.predict_proba(X_test)[:,1])
roc_auc = auc(fpr, tpr)
print('KNN AUC:', roc_auc)

# 逻辑回归的AUC值
fpr, tpr, thresholds = roc_curve(y_test, lr.predict_proba(X_test)[:,1])
roc_auc = auc(fpr, tpr)
print('Logistic Regression AUC:', roc_auc)

# 决策树的AUC值
fpr, tpr, thresholds = roc_curve(y_test, dt.predict_proba(X_test)[:,1])
roc_auc = auc(fpr, tpr)
print('Decision Tree AUC:', roc_auc)
```

```python
from sklearn.model_selection import GridSearchCV

params = {'C': [0.001, 0.01, 0.1, 1, 10, 100]}
grid = GridSearchCV(LogisticRegression(max_iter=2000), params, cv=5, scoring='roc_auc')
grid.fit(X_train, y_train)

print('Best C:', grid.best_params_)
```

```python
from sklearn.feature_selection import RFE

lr = LogisticRegression(max_iter=2000, C=0.1)
selector = RFE(lr, n_features_to_select=10, step=1, verbose=1)
selector.fit(X_train, y_train)

X_train_selected = selector.transform(X_train)
X_test_selected = selector.transform(X_test)

lr.fit(X_train_selected, y_train)
y_pred = lr.predict(X_test_selected)

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
print('Confusion Matrix:\n', confusion_matrix(y_test, y_pred))
print('Classification Report:\n', classification_report(y_test, y_pred))

fpr, tpr, thresholds = roc_curve(y_test, lr.predict_proba(X_test_selected)[:,1])
roc_auc = auc(fpr, tpr)
print('AUC:', roc_auc)
```
